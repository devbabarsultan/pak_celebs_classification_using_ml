```python
import os
from icrawler.builtin import BingImageCrawler

def scraper(celeb_name):
    bing_crawler = BingImageCrawler(storage={'root_dir': './dataset/'+celeb_name})
    bing_crawler.crawl(keyword=celeb_name, max_num=250,overwrite=False)
    
```

```python
political_leaders = ['Imran_Khan','Atif_Aslam','Humayun_Saeed','Waqar_Zaqa','Bhola_Record']
```

```python
for leader in political_leaders:
    print(f'Scraping images for {leader}...')
    scraper(leader)
    print(f'Finished scraping images for {leader}.\n')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import requests
import os

# Setup
search_query = "Bhola Record"
folder = "./dataset/" + search_query.replace(" ", "_")
os.makedirs(folder, exist_ok=True)

driver = webdriver.Chrome()
driver.get(f"https://www.google.com/search?tbm=isch&q={search_query}")

# Scroll down to load more images
for _ in range(3):
    driver.execute_script("window.scrollBy(0,1000)")
    time.sleep(1)

images = driver.find_elements(By.TAG_NAME, "img")

count = 0
for img in images:
    try:
        src = img.get_attribute('src')
        if src and "http" in src:
            img_data = requests.get(src).content
            with open(f"{folder}/{search_query}_{count}.jpg", "wb") as f:
                f.write(img_data)
            count += 1
    except:
        continue

driver.quit()
```
